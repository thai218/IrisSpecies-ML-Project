```python
# Import các thư viện cần thiết
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Tải dataset iris từ seaborn
df = sns.load_dataset("iris")

# Hiển thị 5 dòng đầu
print("5 dòng đầu dữ liệu")
display(df.head())

# Thông tin tổng quát (số dòng, cột, dtype)
print("\n Thông tin tổng quát")
df.info()

# Thống kê mô tả (mean, std, min, max…)
print("\n Mô tả dữ liệu")
display(df.describe())

# Phân phối biến mục tiêu
print("\n Phân phối biến mục tiêu")
print(df['species'].value_counts())
```

```python
# Kiểm tra kiểu dữ liệu
print("\nKiểu dữ liệu:")
print(df.dtypes)
```

```python
# Kiểm tra giá trị thiếu
print("Missing values:")
print(df.isnull().sum())
```

```python
# Boxplot từng feature phân theo species
features = df.columns[:-1]  # tất cả trừ 'species'
for col in features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f"Boxplot của {col} theo loài")
    plt.xlabel("Species")
    plt.ylabel(col)
    plt.show()
```

```python
# Histograms cho mỗi feature
df.hist(figsize=(10, 8), bins=20)
plt.suptitle("Histogram cho các đặc trưng")
plt.show()
```

```python
# Pairplot để quan sát mối quan hệ 2 chiều và phân phối
sns.pairplot(df, hue="species", diag_kind="kde")
plt.suptitle("Pairplot các đặc trưng theo loài")
plt.show()
```

```python
# Tính correlation matrix (chỉ với các cột số)
corr = df.drop('species', axis=1).corr()

# Vẽ heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```
